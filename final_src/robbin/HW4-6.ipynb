#### $V=\pi h^2 \frac{3R-h}{3}$, substitute R=3, V=30,  find the root of the function: $f(h)=\pi h^2 \frac{3R-h}{3}-V=0 $

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### $f(h)=-\frac{\pi h^3}{3}+3\pi h^2 -30$
#### $\Rightarrow f'(h)=-\pi h^2 + 6\pi h$

```python
# data
ea = 1.0  # approximate relative error

x_previous = 0.0
x_current = 2.0

iteration_time = 3
```

```python
# function definition
def func(x):
    return -(np.pi * x**3) / 3 + 3 * np.pi * x**2 - 30

def func_diff(x):
    return -np.pi * x**2 + 6 * np.pi * x

def func_diff2(x):
    return -2 * np.pi * x + 6 * np.pi
```

```python
# evaluate
for iteration_count in range(iteration_time):
    x_previous = x_current
    #--------------------------- a new iteration
    x_current = x_previous - func(x_previous) / func_diff(x_previous) # Newton-Raphson method
    ea = abs((x_current - x_previous) / x_current)
    print("iteration count: {}, approximate value: {:>6f}, approximate relative error: {:>6f}".format(iteration_count + 1, x_current, ea))
```

```python
# define data points of x-axis and y-axis
# x1, y1 -> y = f(x) = math.exp(-0.5 * x) * (4 - x) - 2
# x2, y2 -> y = f''(x)

x1 = np.linspace(-20, 30, 300)
y1 = func(x1)

x2 = np.linspace(-20, 30, 300)
y2 = func_diff(x1)

fig = plt.figure()

axes = fig.add_subplot(111)

func1, = axes.plot(x1, y1, label="f(h)")
func2, = axes.plot(x2, y2, label="f'(h)")

plt.xlabel("height", fontsize=15)
plt.ylabel("Volume", fontsize=15)

plt.legend(handles=[func1, func2], loc="best")

plt.tick_params(axis="both", labelsize=12, color="red")
plt.axis([-20, 30, -40, 90]) #x-axis show -1~10, y-axis show -3~10
plt.grid()

plt.show()
```

#### h 只有在被$f'(h)$的二根$x_a, x_b$包圍起來的區域才求得到解，$x_a, x_b$為$f(h)$的反曲點
