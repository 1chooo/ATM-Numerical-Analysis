#### Use the Newton-Raphson method to find the root of $f(x) = e^{-0.5x}(4-x)-2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### $f(x) = e^{-0.5x}(4-x)-2$
#### $\Rightarrow f'(x)=-0.5e^{-0.5x}(4-x)-e^{-0.5x}$

```python
# data
es = 0.02 # tolerance
ea = 1.0  # approximate relative error

x_previous = 0.0
x_current = int(input("Choose the initial guess of x"))

iteration_time = 0
```

```python
# function definition
def func(x):
    return np.exp(-0.5 * x) * (4 - x) - 2

def func_diff(x):
    return -0.5 * np.exp(-0.5 * x) * (4 - x) - np.exp(-0.5 * x)
```

```python
# evaluate
while ea > es:
    x_previous = x_current
    #--------------------------- a new iteration
    iteration_time += 1
    x_current = x_previous - func(x_previous) / func_diff(x_previous) # Newton-Raphson method
    ea = abs((x_current - x_previous) / x_current)

print("iteration_time: %d" % iteration_time)
print("approximate value: %f" % x_current)
```

```python
# define data points of x-axis and y-axis
# x1, y1 -> y = f(x) = math.exp(-0.5 * x) * (4 - x) - 2
# x2, y2 -> tangent of f(x) at x = 2
# x3, y3 -> tangent of f(x) at x = 6
# x4, y4 -> tangent of f(x) at x = 8

x1 = np.linspace(-3, 10, 100)
y1 = func(x1)

x2 = np.linspace(-3, 10, 100)
y2 = func_diff(2) * (x2 - 2) + func(2)

x3 = np.linspace(-3, 10, 100)
y3 = func_diff(6) * (x2 - 6) + func(6)

x4 = np.linspace(-3, 10, 100)
y4 = func_diff(8) * (x2 - 8) + func(8)

# plot
fig = plt.figure()

#---------------------------
axes1 = fig.add_subplot(311)

func1, = axes1.plot(x1, y1, label="f(x)")
func2, = axes1.plot(x2, y2, label="tangent at x=2")

plt.xlabel("x", fontsize=15)
plt.ylabel("y", fontsize=15)

plt.legend(handles=[func1, func2], loc="best")

plt.tick_params(axis="both", labelsize=12, color="red")
plt.axis([-3, 10, -3, 10]) #x-axis show -1~10, y-axis show -3~10

#---------------------------
axes2 = fig.add_subplot(312)

func1, = axes2.plot(x1, y1, label="f(x)")
func3, = axes2.plot(x3, y3, label="tangent at x=6")

plt.xlabel("x", fontsize=15)
plt.ylabel("y", fontsize=15)

plt.legend(handles=[func1, func3], loc="best")

plt.tick_params(axis="both", labelsize=12, color="red")
plt.axis([-3, 10, -3, 10]) #x-axis show -1~10, y-axis show -3~10

#---------------------------
axes3 = fig.add_subplot(313)

func1, = axes3.plot(x1, y1, label="f(x)")
func4, = axes3.plot(x4, y4, label="tangent at x=8")

plt.xlabel("x", fontsize=15)
plt.ylabel("y", fontsize=15)

plt.legend(handles=[func1, func4], loc="best")

plt.tick_params(axis="both", labelsize=12, color="red")
plt.axis([-3, 10, -3, 10]) #x-axis show -1~10, y-axis show -3~10

#---------------------------
plt.show()
```
