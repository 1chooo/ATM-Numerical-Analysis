SUB Gauss(a, b, n, x, tol, er)
    DIMENSION s(n)
    er = 0
    DOFOR i = 1, n
        si = ABS(a[i][1])
        DOFOR j = 2, n
            IF ABS(a[i][j]) > si THEN
                si = ABS(a[i][j])
        ENDDO
    ENDDO

END Gauss
